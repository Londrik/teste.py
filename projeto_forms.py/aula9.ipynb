```python
# Lista de vendas dos funcionários
vendas_funcionarios = [2800, 3200, 1500, 4000, 5000, 1000, 3500]

# Meta de vendas para receber o bônus
meta = float(3000)

# Percentual do bônus (5%)
percentual_bonus = 0.05

print("Calculando bônus dos funcionários:")

# Percorre a lista de vendas utilizando um loop for
for i, venda in enumerate(vendas_funcionarios):
    if venda >= meta:
        bonus = venda * percentual_bonus
        print(f"Funcionário {i+1}: Vendas = R$ {venda:.2f}, Bônus = R$ {bonus:.2f}")
    else:
        bonus = 0.00
        print(f"Funcionário {i+1}: Vendas = R$ {venda:.2f}, Bônus = R$ {bonus:.2f} (Meta não atingida)")
```

```python
# Lista de vendas diárias
vendas_diarias = [2000, 3000, 2500, 4000, 1800, 5000, 2700]

# Meta de vendas
meta = float(2500)

# Inicializa um contador para os dias com vendas acima da meta
contador_dias_acima_meta = 0

print("Verificando vendas diárias:")

# Percorre a lista de vendas diárias
for venda in vendas_diarias:
    if venda > meta:
        contador_dias_acima_meta += 1

print(f"Número de dias com vendas acima da meta de R$ {meta:.2f}: {contador_dias_acima_meta}")
```

```python
vendas_vendedores =[1500,2500,1800,3200,4000,1200,2800]
meta = float(2000)
comissao = 0.08 # Corrigido para 0.08 para representar 8%

print("Calculando comissões dos vendedores:")
for venda in vendas_vendedores:
    if venda >= meta:
        comissao_vendedor = venda * comissao
        print(f"Vendedor: R$ {comissao_vendedor:.2f}")
    else: # Indentação do else corrigida
        print("Vendedor: R$ 0.00")
```

```python
lista_vendas = [1000, 500, 800, 1500, 2000, 2300]
meta = float(1200)

vendas_acima_meta = 0

print("Contando vendas que atingiram a meta:")

for venda in lista_vendas:
  if venda >= meta:
    vendas_acima_meta += 1

print(f"Total de vendas que atingiram ou superaram a meta de R$ {meta:.2f}: {vendas_acima_meta}")
```

```python
lista_vendas = [1000, 500, 800, 1500, 2000, 2300]
meta = float(1200)
percentual_bonus = 0.1 # Corrigido para 10%
bonus_extra = 50

print("Calculando bônus com bônus extra para grandes vendas:")

for venda in lista_vendas:
  bonus_total = 0
  if venda >= meta:
    bonus_total = venda * percentual_bonus
    if venda >= 2000:
      bonus_total += bonus_extra
    print(f"Venda = R$ {venda:.2f}, Bônus = R$ {bonus_total:.2f}")
  else:
    print(f"Venda = R$ {venda:.2f}, Bônus = R$ {bonus_total:.2f} (Meta não atingida)")
```

```python
#Inicializa maior_valor e menor_valor: maior_valor é inicializado como 0 (assumindo que todas as vendas são positivas) e menor_valor também como 0, com uma condição especial para a primeira iteração.
#Percorre a lista de vendas: Utiliza um loop for para iterar sobre cada venda na lista_vendas.
#Encontra o maior valor: Se a venda atual for maior que o maior_valor registrado até então, maior_valor é atualizado.
#Encontra o menor valor: Se a venda atual for menor que o menor_valor registrado até então, ou se menor_valor ainda for 0 (o que garante que a primeira venda inicialize menor_valor corretamente), menor_valor é atualizado.Exibe os resultados: Após percorrer toda a lista,ele imprime o maior_valor e o menor_valor encontrados, formatados como moeda
```

```python
lista_vendas = [1000,500,800,1500,2000,2300]
maior_valor = 0
menor_valor = 0

for venda in lista_vendas:
    if venda > maior_valor:
        maior_valor = venda
    if venda < menor_valor or menor_valor == 0:
        menor_valor = venda

print(f"Maior valor de venda: R$ {maior_valor:.2f}")
print(f"Menor valor de venda: R$ {menor_valor:.2f}")
```
