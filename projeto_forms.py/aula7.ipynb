```python
# Solicita a idade e converte para número inteiro
try:
    idade = int(input("Digite a sua idade: "))

    # Classificação por faixas
    if idade < 0:
        print("Idade inválida.")
    elif idade <= 12:
        print("Criança")
    elif idade <= 17:
        print("Adolescente")
    elif idade <= 59:
        print("Adulto")
    else:
        print("Idoso")

except ValueError:
    print("Por favor, digite um número válido.")
```
