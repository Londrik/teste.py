````python
#python
# Exemplo de if-else

idade = 18

if idade >= 18:
    print("Você é maior de idade.")
else:
    print("Você é menor de idade.")

# Tente mudar o valor da variável 'idade' e execute a célula novamente!
idade = 16

if idade >= 18:
    print("Você é maior de idade.")
else:
    print("Você é menor de idade.")
````

```markdown
## Como funciona:

*   A instrução `if` verifica se uma condição é verdadeira.
*   Se a condição for `True` (verdadeira), o bloco de código indentado abaixo do `if` é executado.
*   Se a condição for `False` (falsa), o bloco de código indentado abaixo do `else` é executado.

Neste exemplo, estamos verificando se a `idade` é maior ou igual a 18.

```markdown
## Desafio: Par ou Ímpar?

Seu desafio é criar um código que verifica se um número inteiro é par ou ímpar.

*   **Instruções:**
    1.  Crie uma variável `numero` e atribua a ela um valor inteiro.
    2.  Use uma instrução `if-else` para verificar se o `numero` é par ou ímpar.
    3.  Imprima uma mensagem indicando se o número é "Par" ou "Ímpar".

**Dica:** O operador de módulo (`%`) retorna o resto de uma divisão. Um número par dividido por 2 tem resto 0.
```

````python
#python
# Solução para o desafio 'Par ou Ímpar'

numero = 7 # Tente mudar este valor para testar!

if numero % 2 == 0:
    print(f"O número {numero} é Par.")
else:
    print(f"O número {numero} é Ímpar.")


numero = 10 # Outro exemplo

if numero % 2 == 0:
    print(f"O número {numero} é Par.")
else:
    print(f"O número {numero} é Ímpar.")
````

```markdown
## Explicação da Solução:

Nesta solução, utilizamos o operador de módulo (`%`).

*   `numero % 2` calcula o resto da divisão do `numero` por 2.
*   Se o resto for `0`, significa que o número é divisível por 2, portanto, é **Par**.
*   Se o resto não for `0` (ou seja, for `1` para números inteiros), o número não é divisível por 2, então é **Ímpar**.

O `if-else` direciona o fluxo do programa para a mensagem correta com base no resultado dessa verificação.
```

```markdown
## Desafio: Positivo, Negativo ou Zero?

Crie um código que determine se um número inteiro é positivo, negativo ou zero.

*   **Instruções:**
    1.  Crie uma variável `valor` e atribua a ela um número inteiro.
    2.  Use instruções `if`, `elif` e `else` para verificar as condições:
        *   Se `valor` for maior que zero, imprima "O número é Positivo."
        *   Se `valor` for menor que zero, imprima "O número é Negativo."
        *   Caso contrário (se for zero), imprima "O número é Zero."

**Dica:** Lembre-se de que `elif` é usado para verificar condições adicionais após um `if`.
```

````python
#```python
# Escreva seu código para o desafio 'Positivo, Negativo ou Zero' aqui!

valor = int(input("Digite um numero inteiro: "))

if valor > 0:
    print("O numero é positivo")
elif valor < 0:
    print("O numero é negativo")
else:
    print("O numero é zero")




````

## Desafio: Divisível por 3 e 5?

Crie um código que determine se um número inteiro é divisível por 3, por 5, por ambos, ou por nenhum deles.

*   **Instruções:**
    1.  Crie uma variável `numero` e atribua a ela um valor inteiro.
    2.  Use instruções `if`, `elif` e `else` para verificar as condições:
        *   Se `numero` for divisível por 3 e por 5, imprima "Divisível por 3 e 5."
        *   Se `numero` for divisível apenas por 3, imprima "Divisível apenas por 3."
        *   Se `numero` for divisível apenas por 5, imprima "Divisível apenas por 5."
        *   Caso contrário, imprima "Não é divisível por 3 nem por 5."

**Dica:** Use o operador de módulo (`%`) e o operador lógico `and` para combinar condições.
```

````python
#python
# Escreva seu código para o desafio 'Divisível por 3 e 5?' aqui!

numero = int(input("digite um numero:"))

if numero % 3 == 0 and numero % 5 == 0:
    print("Divisível por 3 e 5")
elif numero % 3 == 0:
    print("Divisível apenas por 3")
elif numero % 5 == 0:
    print("Divisível apenas por 5")
````
