## Tratamento de Dados de Usuário

Este notebook processa dados de nome e e-mail de um usuário.

```python
# Dados fornecidos
nome = " MARIA EDUARDA ALMEIDA "
email = " MARIA.ALMEIDA@EMPRESA.COM "

print(f"Nome original: '{nome}'")
print(f"Email original: '{email}'")
```

### 1- Remover espaços extras

```python
nome_sem_espacos = nome.strip()
email_sem_espacos = email.replace(" ", "") # Remover todos os espaços do email

print(f"Nome sem espaços extras: '{nome_sem_espacos}'")
print(f"Email sem espaços extras: '{email_sem_espacos}'")
```

### 2- Converter para minúsculo

```python
nome_minusculo = nome_sem_espacos.lower()
email_minusculo = email_sem_espacos.lower()

print(f"Nome em minúsculo: '{nome_minusculo}'")
print(f"Email em minúsculo: '{email_minusculo}'")
```

### 3- Substituir domínio do email

```python
email_novo_dominio = email_minusculo.replace("empresa.com", "gmail.com")

print(f"Email com novo domínio: '{email_novo_dominio}'")
```

### 4- Mostrar quantidade de caracteres do nome (sem espaços extras)

```python
quantidade_caracteres_nome = len(nome_sem_espacos)

print(f"Quantidade de caracteres do nome (sem espaços extras): {quantidade_caracteres_nome}")
```

### 5- Exibir a primeira e a última letra do nome

```python
primeira_letra_nome = nome_sem_espacos[0]
ultima_letra_nome = nome_sem_espacos[-1]

print(f"Primeira letra do nome: '{primeira_letra_nome}'")
print(f"Última letra do nome: '{ultima_letra_nome}'")
```

## ATIVIDADE — Tratamento de Dados de Produto

Este notebook processa dados de um produto.

```python
# Dados fornecidos do produto
produto = " NOTEBOOK DELL INSPIRON "
categoria = " INFORMATICA "
codigo = " ABC-123-XYZ "

print(f"Produto original: '{produto}'")
print(f"Categoria original: '{categoria}'")
print(f"Código original: '{codigo}'")
```

### 1- Remover espaços extras de todas as variáveis

```python
produto_sem_espacos = produto.strip()
categoria_sem_espacos = categoria.strip()
codigo_sem_espacos = codigo.replace(" ", "") # Remove todos os espaços do código

print(f"Produto sem espaços extras: '{produto_sem_espacos}'")
print(f"Categoria sem espaços extras: '{categoria_sem_espacos}'")
print(f"Código sem espaços extras: '{codigo_sem_espacos}'")
```

### 2- Converter a categoria para minúsculo

```python
categoria_minuscula = categoria_sem_espacos.lower()

print(f"Categoria em minúsculo: '{categoria_minuscula}'")
```

### 3- Padronizar o código do produto (maiúsculas e sem traços)

```python
codigo_padronizado = codigo_sem_espacos.upper().replace("-", "")

print(f"Código padronizado: '{codigo_padronizado}'")
```

### 4- Mostrar quantidade de caracteres do nome do produto (sem espaços extras)

```python
quantidade_caracteres_produto = len(produto_sem_espacos)

print(f"Quantidade de caracteres do nome do produto (sem espaços extras): {quantidade_caracteres_produto}")
```

### 5- Exibir a primeira letra do nome do produto e os 4 primeiros caracteres do código

```python
primeira_letra_produto = produto_sem_espacos[0]
primeiros_4_caracteres_codigo = codigo_padronizado[0:4]

print(f"Primeira letra do nome do produto: '{primeira_letra_produto}'")
print(f"Os 4 primeiros caracteres do código do produto: '{primeiros_4_caracteres_codigo}'")
```
