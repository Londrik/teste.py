```python
numero = float(input("Digite um número: ")) # Solicita ao usuário que digite um número e o converte para ponto flutuante.

# Verifica se o número é maior que zero
if numero > 0:
    print("O número é positivo.") # Imprime que o número é positivo
# Caso contrário, verifica se o número é menor que zero
elif numero < 0:
    print("O número é negativo.") # Imprime que o número é negativo
# Se não for nem maior nem menor que zero, é zero
else:
    print("O número é zero.") # Imprime que o número é zero.
```

```python
numero = int(input("Digite um número inteiro: ")) # Solicita ao usuário que digite um número inteiro e o converte para inteiro.

# Verifica se o resto da divisão do número por 2 é igual a 0. Se for, o número é par.
if numero % 2 == 0:
    print(f"O número {numero} é par.") # Imprime que o número é par
# Caso contrário, o número é ímpar.
else:
    print(f"O número {numero} é ímpar.") # Imprime que o número é ímpar.
```

```python
nota = float(input("Digite a nota do aluno (entre 0 e 10): ")) # Solicita a nota do aluno e a converte para ponto flutuante.

# Verifica se a nota é maior ou igual a 7
if nota >= 7:
  print(f"A nota {nota} indica que o aluno está aprovado.") # Indica que o aluno está aprovado
# Caso contrário, verifica se a nota é maior ou igual a 5
elif nota >= 5:
  print(f"A nota {nota} indica que o aluno está em recuperação.") # Indica que o aluno está em recuperação
# Se a nota for menor que 5
else:
  print(f"A nota {nota} indica que o aluno está reprovado.") # Indica que o aluno está reprovado.
```

```python
usuario = input("Digite o nome de usuário: ") # Solicita o nome de usuário
senha = input("Digite a senha: ") # Solicita a senha

# Verifica se o nome de usuário é 'admin' E a senha é 'admin123'
if usuario == "admin" and senha == "admin123":
  print("Login bem-sucedido!") # Imprime mensagem de login bem-sucedido
# Caso contrário (se o usuário ou a senha estiverem incorretos)
else:
  print("Nome de usuário ou senha incorretos") # Imprime mensagem de erro de login
```
