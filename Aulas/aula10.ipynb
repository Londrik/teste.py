```python
contador = 10

while contador >= 0:
    print(contador)
    contador -= 1

print("Fim da contagem!")
```

```python
numero = 0

while numero <= 20:
    if numero % 2 == 0:
        print(numero)
    numero += 1
```

```python
soma = 0
numero_digitado = 0  # Inicializa com um valor diferente de -5 para entrar no loop

while numero_digitado != -5:
    numero_digitado = int(input("Digite um número inteiro (-5 para parar): "))
    if numero_digitado != -5:
        soma += numero_digitado

print(f"A soma dos números digitados (excluindo -5) é: {soma}")
```

```python
senha_correta = "cofre2025"
senha_digitada = "" # Inicializa com uma senha incorreta para entrar no loop

while senha_digitada != senha_correta:
    senha_digitada = input("Digite o código do cofre: ")
    if senha_digitada != senha_correta:
        print("Código incorreto. Tente novamente.")

print("Cofre aberto com sucesso!")
```
