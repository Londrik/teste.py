```python
idade1 = int(input("Digite a idade da primeira pessoa: "))
idade2 = int(input("Digite a idade da segunda pessoa: "))

print(f"{idade1} é igual a {idade2}? {idade1 == idade2}")
print(f"{idade1} é diferente de {idade2}? {idade1 != idade2}")
print(f"{idade1} é maior que {idade2}? {idade1 > idade2}")
print(f"{idade1} é maior ou igual a {idade2}? {idade1 >= idade2}")
print(f"{idade1} é menor que {idade2}? {idade1 < idade2}")
print(f"{idade1} é menor ou igual a {idade2}? {idade1 <= idade2}")
```

```python
# Entrada de dados
is_vip = input("O cliente é VIP? (sim/nao): ").strip().lower()
valor_compra = float(input("Qual o valor da compra? R$"))

# O cliente ganha desconto se for VIP OU se o valor for maior que 100
ganha_desconto = (is_vip == "sim") or (valor_compra > 100)

# Mostra o resultado da verificação
print(ganha_desconto)
```

```python
# Recebe o dia do usuário
dia = input("Digite o dia da semana: ").strip().lower()

# Lista de dias úteis
dias_uteis = ["segunda", "terça", "quarta", "quinta", "sexta"]

# A lógica: Se o dia NÃO estiver na lista de dias úteis
if dia not in dias_uteis:
    print("A pessoa está de descanso")
else:
    print("A pessoa está em dia útil")
```
